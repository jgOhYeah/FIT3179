## Download and extract the data

The web portal will only allow 180 stations to be downloaded at a time. This means that many downloads need to be made to cover the entirity of Victoria. These then need to be merged back into a single dataset.

```python
# %%bash
# urls.txt is a text file containing one emailed download link per line.
# wget -i urls.txt
```

```python
%%bash
# Extract each zip file.
for filename in *.zip; do
    echo "$filename"
    folder="${filename}_extracted"
    mkdir "$folder"
    unzip "$filename" -d "$folder"
done
```

## Merge the site details files

```python
%%bash
# Get rid of any old file and create the folder to place the merged data in.
rm -rf combined
mkdir -p combined
for dirname in *.zip_extracted; do
    sed "1d" "$dirname/Site Details.csv" >> "combined/Site Details with duplicates.csv"
done
```

```python
%%bash
echo "Number of lines including duplicates"
cat "combined/Site Details with duplicates.csv" | wc
```

```python
%%bash
# Copy the header into the new file
header_file=$(find *.zip_extracted -print -quit)
header=$(head -n 1 "$header_file/Site Details.csv")
echo "$header" | tee "combined/Site Details.csv"

# Add all unique lines except the header. Sorting is required as unique operates on adjacent lines.
cat "combined/Site Details with duplicates.csv" | sort -u >> "combined/Site Details.csv"
```

```python
%%bash
# Count how much we got
echo "Number of lines without duplicates"
wc "combined/Site Details with duplicates.csv" "combined/Site Details.csv"
```

### Generate a single file containing the maximum water level for each town each day

```python
import glob
import os
import datetime

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

path = "./combined"
all_files = glob.glob(os.path.join(path , "*.MeanWaterLevel.csv"))
```

```python
def parse_date(string):
    if type(string) is str:
        return datetime.datetime.strptime(string, "%H:%M:%S %d/%m/%Y")
    else:
        return np.NaN
```

```python
def pretty_name(name):
    return name.title(
    ).replace(
        "@", "at"
    ).replace(
        "D/S", "downstream of"
    ).replace(
        "U/S", "upstream of"
    ).replace(
        "O/Take", "offtake"
    ).replace(
        "Syphn", "Syphon"
    ).replace(
        "Ch ", "Channel "
    ).replace(
        "Ch.", "Channel "
    ).replace(
        "'S ", "'s "
    )
```

```python
# Import basins
basins = pd.read_csv("../FloodLevels/Basins.csv")
basins
```

```python
frames = []
for file in all_files:
    # file = "233246.MeanWaterLevel.csv"

    station = pd.read_csv(file, skiprows=2)
    station["Date"] = station["Date"].apply(parse_date)
    station["Day"] = station["Date"].dt.floor('d')

    # Aggregate
    daily = station.groupby(["Day"])["Mean"].mean()
    daily = daily.to_frame()

    # Add metadata
    id = file.strip("./combined").rstrip(".MeanWaterLevel.csv")
    id_int = int(id)
    basin_code = int(id[:3])
    info = data[data["Site Id"] == id_int]
    daily["Station"] = pretty_name(info["Site Name"].iloc[0])
    try:
        daily["Basin"] = basins["Basin"][basins["Prefix"] == basin_code].iloc[0]
    except:
        print(f"{basin_code=} is not found")
#     daily["Mean"] -= info["Gauge Zero (Ahd)"].iloc[0]

    frames.append(daily)
```

```python
concatenated = pd.concat(frames, axis=0)
concatenated.to_csv("River_Heights.csv")
```
