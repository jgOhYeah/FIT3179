## Download and extract the data

The web portal will only allow 180 stations to be downloaded at a time. This means that many downloads need to be made to cover the entirity of Victoria. These then need to be merged back into a single dataset.

```python
# %%bash
# urls.txt is a text file containing one emailed download link per line.
# wget -i urls.txt
```

```python
%%bash
# Extract each zip file.
for filename in *.zip; do
    echo "$filename"
    folder="${filename}_extracted"
    mkdir "$folder"
    unzip "$filename" -d "$folder"
done
```

## Merge the site details files

```python
%%bash
# Get rid of any old file and create the folder to place the merged data in.
rm -rf combined
mkdir -p combined
for dirname in *.zip_extracted; do
    sed "1d" "$dirname/Site Details.csv" >> "combined/Site Details with duplicates.csv"
done
```

### Generate a single file containing the maximum water level for each town each day

```python
import glob
import os
import datetime

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

path = "./combined"
all_files = glob.glob(os.path.join(path , "*.MeanWaterLevel.csv"))
```

```python
def parse_date(string):
    if type(string) is str:
        return datetime.datetime.strptime(string, "%H:%M:%S %d/%m/%Y")
    else:
        return np.NaN
```

```python
def pretty_name(name):
    return name.title(
    ).replace(
        "@", "at"
    ).replace(
        "D/S", "downstream of"
    ).replace(
        "U/S", "upstream of"
    ).replace(
        "O/Take", "offtake"
    ).replace(
        "Syphn", "Syphon"
    ).replace(
        "Ch ", "Channel "
    ).replace(
        "Ch.", "Channel "
    ).replace(
        "'S ", "'s "
    )
```

```python
# Import basins
# basins = pd.read_csv("../FloodLevels/Basins.csv")
basins = pd.read_csv("../../FloodLevels/Basins.csv")
basins
```

```python
frames = []
for file in all_files:
    # file = "233246.MeanWaterLevel.csv"

    station = pd.read_csv(file, skiprows=2)
    station["Date"] = station["Date"].apply(parse_date)
    station["Day"] = station["Date"].dt.floor('d')

    # Aggregate
    daily = station.groupby(["Day"])["Mean"].mean()
    daily = daily.to_frame()

    # Add metadata
    id = file.strip("./combined").rstrip(".MeanWaterLevel.csv")
    id_int = int(id)
    basin_code = int(id[:3])
    info = data[data["Site Id"] == id_int]
    daily["Station"] = pretty_name(info["Site Name"].iloc[0])
    try:
        daily["Basin"] = basins["Basin"][basins["Prefix"] == basin_code].iloc[0]
    except:
        print(f"{basin_code=} is not found")
#     daily["Mean"] -= info["Gauge Zero (Ahd)"].iloc[0]

    frames.append(daily)
```

```python
concatenated = pd.concat(frames, axis=0)
concatenated.to_csv("River_Heights.csv")
```

# Generate max height data for each region

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', 50)
```

```python
levels = pd.read_excel("../../FloodLevels/FloodLevels.ods", 1)
levels = levels.set_index("AWRC No. ")
levels
```

```python
details = pd.read_csv("Site Details.csv")
details = details.set_index("Site Id")
details
```

```python
details_levels = levels.join(details, how="inner")
details_levels
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
# setup mercator map projection.
m = Basemap(llcrnrlon=140,llcrnrlat=-39.5,urcrnrlon=151,urcrnrlat=-33,\
            rsphere=(6378137.00,6356752.3142),\
            resolution='i',projection='merc',\
            lat_0=40.,lon_0=-20.,lat_ts=20.)

m.drawcoastlines()
m.fillcontinents()
m.drawstates()
m.drawrivers()

x,y = m(details_levels["Longitude"], details_levels["Latitude"])
m.scatter(x, y)
plt.title("Water monitoring sites in Victoria")
plt.show()
```

```python
def max_height(site_id):
    try:
        heights = pd.read_csv(f"{site_id}.MeanWaterLevel.csv", skiprows=2)
        max_row = heights.loc[heights["Mean"].idxmax()]
    except (KeyError, FileNotFoundError):
        return np.NaN
    else:
        return pd.Series([max_row["Date"], max_row["Mean"]])
```

```python
details_levels[["Max Height Date", "Max Height"]] = details_levels.apply(lambda row: max_height(row.name), axis=1)
```

```python
def categorise(row):
    try:
        major = float(row["Major "])
    except ValueError:
        major = 1000
        
    try:
        moderate = float(row["Moderate "])
    except ValueError:
        moderate = 1000
        
    try:
        minor = float(row["Minor "])
    except ValueError:
        minor = 1000
        
    if row["Max Height"] >= major:
        return "Major"
    elif row["Max Height"] >= moderate:
        return "Moderate"
    elif row["Max Height"] >= minor:
        return "Minor"
    elif major == 1000 and moderate == 1000 and minor == 1000:
        return "Unclassified"
    else:
        return "Below minor"
    
details_levels["MaxCategory"] = details_levels.apply(categorise, axis=1)
```

```python
def correct_zeros(row):
    row["Offset"] = 0
    if(row["Max Height"] > 50):
        # AHD
        datum = float(row["Gauge Zero (Ahd)"])
        row["Offset"] = -datum
        row["Max Height"] -= datum
        
        def update(index):
            try:
                row[index] = float(row[index]) - datum
            except ValueError:
                pass
        
        update("Major ")
        update("Moderate ")
        update("Minor ")
    
    return row

details_levels = details_levels.apply(correct_zeros, axis=1)
```

## Generate a massive CSV with data for all stations over time

```python
details_levels = details_levels[details_levels["Max Height"] < 50]
```

```python
frames = []
def over_time(site_id, name, max_category, major, moderate, minor, offset):
    station = pd.read_csv(f"{site_id}.MeanWaterLevel.csv", skiprows=2)
    station["Date"] = station["Date"].apply(parse_date)
    station["Day"] = station["Date"].dt.floor('d')

    # Aggregate
    daily = station.groupby(["Day"])["Mean"].mean()
    daily = daily.to_frame()
#     daily["Mean"] -= offset
    
    # Metadata
    daily["Site ID"] = site_id
    daily["Name"] = name
    
    frames.append(daily)
    
details_levels.apply(lambda row: over_time(row.name, row["Station "], row["MaxCategory"], row["Major "], row["Moderate "], row["Minor "], row["Offset"]), axis=1)
```

```python
concatenated = pd.concat(frames, axis=0)
concatenated = concatenated[concatenated["Mean"] < 40]
concatenated.to_csv("RiverHeights.csv")
```
