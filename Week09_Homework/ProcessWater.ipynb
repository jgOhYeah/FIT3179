## For a given site, find the maximum height and time of that height
Attempting to normalise by taking the change in height from 01/10/2022.

```python
def max_height(site_id):
    try:
        heights = pd.read_csv(f"{site_id}.MeanWaterLevel.csv", skiprows=2)
        max_row = heights.loc[heights["Mean"].idxmax()]
    except (KeyError, FileNotFoundError):
        return np.NaN
    else:
        initial = heights.loc[0]
        difference = max_row["Mean"] - initial["Mean"]
        return pd.Series([max_row["Date"], difference])
```

```python
# max_height(404222)
```

## Find the maximum heights and dates for all sites

```python
sites[["Max Height Date", "Max Height"]] = sites.apply(lambda row: max_height(row["Site Id"]), axis=1)
```

```python
sites
```

## Make the names pretty

```python
def pretty_name(name):
    return name.title(
    ).replace(
        "@", "at"
    ).replace(
        "D/S", "downstream of"
    ).replace(
        "U/S", "upstream of"
    ).replace(
        "O/Take", "offtake"
    ).replace(
        "Syphn", "Syphon"
    ).replace(
        "Ch ", "Channel "
    ).replace(
        "Ch.", "Channel "
    ).replace(
        "'S ", "'s "
    )
```

```python
sites["Name"] = sites["Site Name"].apply(pretty_name)
```

```python
sites
    
```

## Convert dates to days since the start

```python
def parse_date(string):
    if type(string) is str:
        return datetime.datetime.strptime(string, "%H:%M:%S %d/%m/%Y")
    else:
        return np.NaN
```

```python
start = pd.read_csv("404222.MeanWaterLevel.csv",skiprows=2)["Date"][0]
start_date = parse_date(start)

sites["Date"] = sites["Max Height Date"].apply(parse_date)
```

```python
deltas = sites["Date"] - start_date
sites["Days"] = deltas.apply(lambda d: d.days)
```

```python
sites
```

## Filter out sites without data and remove uncecessary columns

```python
filtered = sites[sites["Max Height"].isna() == False]
```

```python
result = filtered[["Name", "Max Height", "Days"]]
```
